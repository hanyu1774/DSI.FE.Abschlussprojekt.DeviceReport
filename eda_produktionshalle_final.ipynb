# Explorative Datenanalyse – Produktionshallen-Reporting




# Vorwort / Erklärung zum Umgang mit der vorliegenden explorativen Datenanalyse

Ziel der explorative Datenanalyse (EDA):  

Im Wesentlichen versucht diese EDA, dass die interessantesten Informationen der zugrundeliegenden Daten des Produktionshallen-Reportings einfach und verständlich für ein (fiktives) Unternehmen dargestellt werden. Dadurch sollen Einblicke in die Leistungsfähigkeit von Maschinen des Produktionshallen-Reporting-Datensatzes gewonnen werden.  
Auf diesen Erkenntnissen können Unternehmen eine Risikoeinschätzung ihres Produktionsprozesses datenbasiert kalkulieren oder Produktionsprozesse optimieren und somit die Unternehmensumsätze und Effizienz steigern. So könnten sich beispielsweise Handlungsempfehlungen, zur Anschaffung neuerer Maschinen oder die Verbesserung von Reparaturarbeiten, für ein Unternehmen aus den Ergebnissen dieser EDA erschließen.  
Die wohl umfangreichste Fragestellung dieser EDA lautet: „Für welche Maschine(n) sollten die Techniker am ehesten eine Weiterbildung bekommen, um zukünftige Probleme zeiteffizienter zu bearbeiten?“.  
Diese Fragestellung (wie auch andere Fragestellungen und Feststellungen dieser EDA) soll Unternehmen eine datenbasierte Empfehlung geben, um Produktionsprozesse langfristig zu verbessern.

---------

Wie die EDA vorgeht, um Informationen zu vermitteln:  
1) Es wurde versucht die EDA in ihrer Aussagekraft weitestgehend intuitiv verständlich und selbsterklärend zu gestalten. Dafür wurde zum einen eine Struktur gewählt, die immer wieder aufzeigt, um welche Daten es sich gerade handelt, und zum anderen eine Vielzahl von grafischen Darstellungen genutzt.  
2) An einigen Stellen erläutert diese EDA Ergebnisse, welche potenziell einen besonders hohen Mehrwert für ein Unternehmen haben können oder schwieriger zu interpretieren sind. Diese Einträge wurden als Markdown-Texteinträge, durch das Schlagwort "Erkenntnis" eingeleitet. Diese Texteinträge beziehen sich logischerweise ausschließlich auf den hier untersuchten Datensatz. Bei der Verwendung von anderen Datensätzen sind andere Schlussfolgerungen zu ziehen. Ähnliche individuelle Vermerke (für Erläuterungen) wurden mit Nutzung des Schlagwortes "Notiz" durchgeführt.  
3) Die Codeeinträge wurden weitestgehend so geschrieben, dass, bei der Nutzung eines neuen Datensatzes (mit der gleichen Tabellenstruktur), sowohl numerisch als auch grafisch sinnvolle Ergebnisse entstehen. Dadurch soll eine Wiederverwendbarkeit der EDA gewährleistet werden.  
4) An einigen Stellen wurden Outputs von "print()" oder "display()" Befehlen mit Hilfe eines "#" ausgeblendet. Dieses Vorgehen dient dazu, dass die Funktion des Codes ausgeführt werden kann, während der Nutzen des Outputs allerdings eher uninteressant ist und keinen Mehrwert für die EDA generiert. Meistens dienen diese Codeblöcke, mit einem verborgenen Output, für darauf aufbauende Codeblöcke, die eine grafische Darstellung der Daten bieten.
Die Entscheidung: Den Eintrag „#display()“ anzugeben und nicht ganz weg zu lassen, ist sinnvoll damit Nutzende des Notebooks, die besagten Codeblöcke theoretisch schnell ausführen können, um einen genaueren Einblick in die numerischen Daten zu bekommen und das Verständnis der Daten zu verbessern.

---------

Anmerkungen zur Struktur und Inhalt der EDA:  
1) Die relevantesten Analyseerkenntnisse der Datenbank können aus dem Kapitel der Ticketanalyse gezogen werden. 
2) Es ist wichtig zu berücksichtigen, dass Fehler und Fehlercodes nicht das gleiche sind und somit nicht synonym verwendet werden können. An mehreren Stellen kann es hier zur Verwechslungsgefahr kommen.

# 1. Vorbereitung und Überblick der Datenbank für die EDA

## 1.1 Datenbank importieren

```python
from pathlib import Path
import sqlite3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator 
    # um in den grafischen Darstellungen ganzzahlige Werte auf den Achsen zu betrachten (mittels: plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True)))

pd.set_option("display.max_columns", None)

DB_PATH = Path("analytic_database_v4.db")

connection = sqlite3.connect(DB_PATH)
print(f"Verbunden mit: {DB_PATH.resolve()}")
```

## 1.2 Tabellen laden

```python
error_codes = pd.read_sql_query("SELECT * FROM error_codes;", connection)
event_log = pd.read_sql_query("SELECT * FROM event_log;", connection)
halls = pd.read_sql_query("SELECT * FROM halls;", connection)
machine_types = pd.read_sql_query("SELECT * FROM machine_types;", connection)
machines = pd.read_sql_query("SELECT * FROM machines;", connection)
measures = pd.read_sql_query("SELECT * FROM measures;", connection)
technicians = pd.read_sql_query("SELECT * FROM technicians;", connection)
ticket_types = pd.read_sql_query("SELECT * FROM ticket_types;", connection)
tickets = pd.read_sql_query("SELECT * FROM tickets;", connection)

print("error_codes:", error_codes.shape)
print("event_log:", event_log.shape)
print("halls:", halls.shape)
print("machine_types:", machine_types.shape)
print("machines:", machines.shape)
print("measures:", measures.shape)
print("technicians:", technicians.shape)
print("ticket_types:", ticket_types.shape)
print("tickets:", tickets.shape)
```

## 1.3 Überblick der Tabellen

### Angabe der ersten 5 Zeilen zu allen Tabellen der Datenbank

```python
print("error_codes:")
display(error_codes.head())
print()

print("event_log:")
display(event_log.head())
print()

print("halls:")
display(halls.head())
print()

print("machine_types:")
display(machine_types.head())
print()

print("machines:")
display(machines.head())
print()

print("measures:")
display(measures.head())
print()

print("technicians:")
display(technicians.head())
print()

print("ticket_types:")
display(ticket_types.head())
print()

print("tickets:")
display(tickets.head())
print()
```

### Wesentliche Informationen zu den einzelnen Tabellen

```python
print("Angabe der fehlenden Werte und Datentypen.")
print()

tables = {
    "error_codes": error_codes,
    "event_log": event_log,
    "halls": halls,
    "machine_types": machine_types,
    "machines": machines,
    "measures": measures,
    "technicians": technicians,
    "ticket_types": ticket_types,
    "tickets": tickets
}

for name, df in tables.items():
    print("=" * 60)
    print(name)
    print("=" * 60)
    print(f"Zeilen, Spalten: {df.shape}")
    print("\nDatentypen:")
    display(df.dtypes)
    print("\nFehlende Werte:")
    display(df.isnull().sum())
    print()
```

Notiz:

Die fehlenden Werte in der event_log Tabelle kommen daher, dass ein event_log Ereignis kein tatsächlicher Fehler ist, sondern eine "maintenance" Wartung stattfindet oder die Maschine offline ist (siehe event_log status) .


## 1.4 Bereinigung der Datentypen

In einigen Tabellen werden Datumsangaben genutzt, die in jupyter Notebook schwer dargestellt werden können.
Dafür werden hier mehrere Spaltenangaben in Datumsobjekte umgewandelt.

```python
event_log["timestamp"] = pd.to_datetime(event_log["timestamp"], errors="coerce")

for column in ["created_at", "assigned_at", "resolved_at", "closed_at"]:
    tickets[column] = pd.to_datetime(tickets[column], errors="coerce")

measures["start_date"] = pd.to_datetime(measures["start_date"], errors="coerce")

#display(event_log)
```

# 2. Analyse

## 2.1 Allgemeiner Überblick über die Maschinen

Übersichtliche Zusammenfassung der wichtigsten Eigenschaften der Maschinen aus der Datenbank.

```python
machines_overview = (
    machines
    .merge(machine_types, on="machine_type_id")
    .merge(halls, on="hall_id")
)

# für eine saubere Darstellung - ohne den Spalten "hall_id" und "machine_type_id"
machines_overview = machines_overview[[
    "machine_id",
    "machine_name",
    "machine_type_name",
    "hall_name",
    "manufacturer",
    "year_built"
]]

display(machines_overview)
```

### Anzahl Maschinen je Maschinentyp

```python
machine_type_counts = (
    machines_overview["machine_type_name"]
    .value_counts()
    .sort_values(ascending=False) #Sortierung: Damit die Maschinentypen mit den meisten unterschiedlichen Maschinen als erstes angegeben werden.
)

#display(machine_type_counts)
```

```python
plt.figure(figsize=(8,4))

machine_type_counts.plot(kind="bar")

plt.title("Anzahl Maschinen je Maschinentyp")
plt.xlabel("Maschinentyp")
plt.ylabel("Anzahl Maschinen")
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.xticks(rotation=45)
plt.grid(axis="y")

plt.show()
```

### Anzahl Maschinen je Halle

```python
hall_counts = (
    machines_overview["hall_name"]
    .value_counts()
    .sort_index(ascending=True) #Sortierung: Damit die Hallen überichtlich sortiert (alphabetisiert) angegeben werden.
)

#display(hall_counts)
```

```python
plt.figure(figsize=(6,4))

hall_counts.plot(kind="bar")

plt.title("Maschinen pro Halle")
plt.xlabel("Halle")
plt.ylabel("Anzahl Maschinen")
plt.grid(axis="y")

plt.show()
```

### Hersteller der Maschinen

```python
manufacturer_counts = (
    machines_overview["manufacturer"]
    .value_counts()
    .sort_values(ascending=False) #Sortierung: Damit übersichtlich angegeben wird von welchem Hersteller die meisten verschiedenen Machinen kommen.
)

#display(manufacturer_counts)
```

```python
plt.figure(figsize=(8,4))

manufacturer_counts.plot(kind="bar")

plt.title("Maschinen nach Hersteller")
plt.xlabel("Hersteller")
plt.ylabel("Anzahl Maschinen")
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.grid(axis="y")

plt.show()
```

### Baujahre der Maschinen

```python
building_year = (
    machines_overview["year_built"]
    .value_counts() #Gruppierung
    .sort_index(ascending=False) #Sortierung: Damit übersichtlich angegeben wird in welchem Jahr die meisten Maschinen gebaut wurden.
)

#display(building_year)
```

```python
plt.figure(figsize=(8,4))

building_year.plot(kind="bar")

plt.title("Verteilung der Baujahre")
plt.xlabel("Baujahr")
plt.ylabel("Anzahl Maschinen")
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))

plt.show()
```

### Maschinen Fehlercodes

Hintergrund:
Da jeder Fehlercode mit einem Machinentyp in Verbindung steht, kann ermittelt werden, welche Fehler für welche Maschine grundsätzlich möglich sind bzw. in der Datenbank festgestellt wurden.

#### Merge und Datenübersicht 

Benötigte Tabellen:
machines, machine_types, error_codes

```python
machine_error_codes = (
    machines
    .merge(machine_types, on="machine_type_id")
    .merge(error_codes, on="machine_type_id")
)

display(machine_error_codes.head())
```

#### Fehlercodes pro Maschine

```python
possible_errors = (
    machine_error_codes
    .groupby("machine_name")["error_code"]
    .count()
    .sort_index(ascending=True) # Sortierung
)

#display(possible_errors)
```

```python
plt.figure(figsize=(10,5))

possible_errors.plot(kind="bar")

plt.title("Anzahl der Fehlercodes je Maschine")
plt.xlabel("Maschine")
plt.ylabel("Anzahl Fehlercodes")
plt.xticks(rotation=45)
plt.grid(axis="y")

plt.show()
```

#### Art der Fehlercodes pro Maschine

```python
machine_error_list = (
    machine_error_codes
    .groupby("machine_name")["error_code"]
    .apply(list)
    .reset_index()
)

display(machine_error_list)
```

### Ereignisse pro Monat

Gibt an, wie viele Event-Log Einträge pro Monat aufgezeichnet wurden.  
Bei einem längeren Zeitintervall (erweiterter / neuer Datensatz) könnten hier Saisonale Muster abgeleitet werden.

```python
plt.figure(figsize=(8,4))

sns.boxplot(data=event_log["downtime_minutes"])

plt.title("Verteilung der Ausfallzeiten - Boxplot")
plt.ylabel("Anzahl der Event-Log Einträge")

plt.show()
```

#### Erkenntnis zu den Ausfallzeiten:


Wir können somit festhalten, dass die meisten Ausfallzeiten nicht all zu lange bestehen bleiben.  
Die durchschnittliche Ausfallzeit liegt bei etwas mehr als 80 Minuten (80.504583 Minuten).  
Über 50% der Ereignisse konnten unter einer Stunde behoben werden (Der Median lag mit 54.1 bei etwas mehr als 54 Minuten).  
Auch Ausreißer sind festzustellen, wobei der höchste Ausreißer bei 1152.4 Minuten lag (also etwa bei 19,2 Stunden).

#### Für welche Maschine(n) sollten die Techniker am ehesten eine Weiterbildung bekommen, um zukünftige Probleme zeiteffizienter zu bearbeiten?

Diese Fragestellung unterliegt der folgenden Annahme:  
Alle Maschinen sind gleich schwierig zu warten. Einzig das Know-How, wie eine Maschine zu reparieren ist, variiert.

Schritt 1) Benötigte Daten vorbereiten

```python
training_analysis = (
    tickets
    .merge(machines, on="machine_id")
    .merge(technicians, on="technician_id")
)

#display(training_analysis)
```

Schritt 2) Schwellenwerte aufstellen

Notiz:  
Schwellenwerte geben Handlungsempfehlungen (hier Schwerpunkt der Weiterbildung) eine aussagekräftige und datenbasierte Grundlage. Deshalb werden im folgenden einige Schwellenwerte für das weitere vorgehen aufgestellt.

Die Schwellenwerte wurden in Abhängigkeit des untersuchten Datensatzes erstellt, damit auch zukünftige Untersuchungen (auf Basis anderer Datensätze) mittels der verwendeten Schwellenwerte zu sinnvollen Empfehlungen gelangen können.
Dadurch soll auch unter anderem eine Robustheit der Empfehlung geschaffen werden, da Median und Quartile unempfindlich gegenüber Ausreißern sind.

```python
# Kriterien für Schwellenwerte festlegen
## MIN_TECHNICIANS
## Mindestens die Hälfte aller Techniker des Unternehmens haben die Maschine schonmal bearbeitet.
half_of_the_technichians = len(training_analysis['technician_id'].unique())/2 #die Hälfte aller Techniker des Unternehmens

MIN_TECHNICIANS = half_of_the_technichians  #Mittelwert  
    

## MIN_REPAIRS
## Nur Maschinen betrachten, die mindestens so viele Reparaturen besitzen wie der Median aller Maschinen.
## Dient dazu, dass eine Weiterbildung nur dann empfohlen wird, wenn die Fehleranfälligkeit so hoch ist dass übermäßig viele Tickets zustande kommen.
## Dadurch soll die Relevanz der betrachteten Maschine für das Unternehmen gewährleistet werden.
repairs_per_machine = (
    training_analysis
    .groupby("machine_name")
    .size()
) 

MIN_REPAIRS = repairs_per_machine.median()   #Median       


## MIN_TECHNICIANS und MIN_REPAIRS sind Voraussetzungen für eine Maschine, dass sie ausreichend Daten hat um als Schulungsempfehlung genannt zu werden:


## REPAIR_THRESHOLD
## dient dazu einen Schwellenwert für eine Weiterbildung zu empfehlen.
## Ist ein Schwellenwert für die durchschnittliche Reparaturdauer in Stunden

#Version 1
REPAIR_THRESHOLD_1 = (
    training_analysis["repair_hours"]
    .quantile(0.75)
) #75% Quantil (oberes Quartil)


#Version 2
REPAIR_THRESHOLD_2 = (
    training_analysis["repair_hours"].mean()
 #   +
  #  training_analysis["repair_hours"].std()
) # Mittelwert
```

Schritt 3) Kennzahlen pro Maschine ermitteln und darstellen

```python
machine_training = (
    training_analysis
    .groupby("machine_name")
    .agg(
        Reparaturen=("ticket_id", "count"),
        Techniker=("technician_id", "nunique"),
        Durchschnittliche_Reparaturdauer=("repair_hours", "mean"),
        Std_Abweichung=("repair_hours", "std")
    )
)

machine_training["Durchschnittliche_Reparaturdauer"] = (
    machine_training["Durchschnittliche_Reparaturdauer"].round(2)
)

machine_training["Std_Abweichung"] = (
    machine_training["Std_Abweichung"].round(2)
)

display(machine_training.head())
```

Schritt 4) Schulungsempfehlung aufstellen

```python
# Schwellenwert Version 1: durchschnittliche Reperaturdauer = 75% Quantil (REPAIR_THRESHOLD_1)

training_candidates_1 = machine_training[
    (machine_training["Techniker"] >= MIN_TECHNICIANS) &
    (machine_training["Reparaturen"] >= MIN_REPAIRS) &
    (machine_training["Durchschnittliche_Reparaturdauer"] >= REPAIR_THRESHOLD_1)
].sort_values(
    by="Durchschnittliche_Reparaturdauer",
    ascending=False
)

#display(training_candidates_1)
```

```python
# Schwellenwert Version 1: durchschnittliche Reperaturdauer = 75% Quantil (REPAIR_THRESHOLD_1)

# ohne dem REPAIR_THRESHOLD_1 (um zu erkennen, welche Maschinen durch diesen gewählten Schwellenwert wegfallen würden)
# ====================================================================================================================
training_candidates_1_1 = machine_training[
    (machine_training["Techniker"] >= MIN_TECHNICIANS) &
    (machine_training["Reparaturen"] >= MIN_REPAIRS) &
    (machine_training["Durchschnittliche_Reparaturdauer"] >= REPAIR_THRESHOLD_1)
].sort_values(
    by="Durchschnittliche_Reparaturdauer",
    ascending=False
)

#display(training_candidates_1_1)
```

```python
# Schwellenwert Version 1: durchschnittliche Reperaturdauer = 75% Quantil (REPAIR_THRESHOLD_1)
plt.figure(figsize=(12,5))

plt.bar(
    training_candidates_1_1.index,
    training_candidates_1_1["Durchschnittliche_Reparaturdauer"]
)

plt.axhline(
    REPAIR_THRESHOLD_1,
    color="red",
    linestyle="--",
    label=f"Schwellenwert ({REPAIR_THRESHOLD_1} h)"
)

plt.title("Schulungsbedarf bei folgenden Maschinen (Schwellenwert Version 1)")
plt.xlabel("Maschine")
plt.ylabel("Durchschnittliche Reparaturdauer [h]")

plt.xticks(rotation=45)

plt.legend()

plt.grid(axis="y")

plt.show()
```

```python
# Schwellenwert Version 2: durchschnittliche Reperaturdauer = Mittelwert (REPAIR_THRESHOLD_2)

training_candidates_2 = machine_training[
    (machine_training["Techniker"] >= MIN_TECHNICIANS) &
    (machine_training["Reparaturen"] >= MIN_REPAIRS) &
    (machine_training["Durchschnittliche_Reparaturdauer"] >= REPAIR_THRESHOLD_2)
].sort_values(
    by="Durchschnittliche_Reparaturdauer",
    ascending=False
)

display(training_candidates_2) #Ergebnis wird gezeigt, da an der Grafik nur schwer zu erkennen ist, ob auch die dritte Maschine den Schwellenwert übersteigt
```

Die Fragestellung, für welche Maschine die Techniker sich weiterbilden sollten, unterlag einige Annahmen.
Die erste Annahme war die Hypothese, dass die Wartungsarbeit für alle Maschinen gleich schwierig ist und lediglich das Wissen der Techniker über die Maschinen variieren kann.
Außerdem wurden Schwellenwerte aufgestellt um die Fragestellung datengetrieben zu untersuchen und zwei Fallunterscheidungen durchzuführen. 

Der Schwellenwert, wie viele Techniker mindestens an der Maschine gearbeitet haben, und der Schwellenwert, dass die betrachtete Maschine mehr Wartungsarbeiten erlebt hat als der Medianwert betrug, wurden als Ausschlusskriterien gewählt. Dadurch sollte die Relevanz für eine Maschinen-Weiterbildung aufgrundlage von Fehleranfälligkeit gewährleistet werden.

Die darauf aufbauende Fallunterscheidung von Version 1 und 2 bietet die Möglichkeit sowohl für linksschiefe als auch für rechtschiefe Datensätze Handlungsempfehlungen zu geben. (Ergebnisse die über dem Schwellenwert des 75% Quantils liegen, liegen auch über dem Median.) Eine Berücksichtigung der Ergebnisse beider Versionen führt zu einer qualifizierteren Handlungsempfehlung.
